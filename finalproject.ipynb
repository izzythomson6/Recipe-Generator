```python
 #Team Python Project- Recipe finder :)
#We use an input function within a defining function and formatted strings to determine users name and their bootcamp 
''' Ask user's information to welcome '''
def name():
    name = input("Please enter your name:  ")
    bootcamp = input("What is the name of your coding bootcamp?: ")
    print ("Hello {} and everyone joining today from {}".format(name.title(),bootcamp.title()))
name()

#Step two import requests for the recipe finder
import requests
''' Retrieve the outcome from API request '''
def recipe_finder(ingredient):
    app_id = "17c63c09"
    app_key = "04f204e3b5bf9c67f50d8319665bc71b"
    recipe_url = 'https://api.edamam.com/search?q={}&app_id={}&app_key={}'.format(ingredient, app_id, app_key)
    result = requests.get(recipe_url)
    stored_recipe = result.json()
    return stored_recipe['hits']

#Step three, display the recipe for each search result
''' Retrieve the label and URI from recipe's results'''
def recipe_results():
    ''' Ask user the name of the ingredient to look for '''
    ingredient = input('Enter an ingredient you wish to use in a recipe: ')

    results = recipe_finder(ingredient)
    
    #Check if there are results for the input recipe
    if (len(results) >0):
        for result in results:
            recipe = result['recipe']
            print("Name: {}".format(recipe['label']))
            print("URI: {}".format(recipe['uri']))
            print()
    else:
        print("It seems your ingredient cannot be found. Please try again...")
        recipe_results()


#Step four, we use if statements to determine if user wants to make a recipe
#this is after the above functions as recipe_finder(ingredient) and recipe_results() must be defined before being called
#if statements are within a function, we make sure to call this at the end of if statements
def search():
    make_a_recipe = input("Do you want to make a meal?y/n: ")
    if make_a_recipe.lower() == "y":
        print("Start thinking about which ingredient you wish to include ...")
        recipe_results()
    elif make_a_recipe.lower() == "n":
        print("Oh right, see you next time.")
    else:
        print("Invalid response, please try again")
search()
        
#Step five, we then use random and lists if user can not think of an ingredient
thinking_of_ingredient= input('Can you think of an ingredient:y/n ')
if thinking_of_ingredient == "n":
   print('If you can not think of an ingredient, or there is just too much choice, we will generate you a random one from this list')
   import random
   random_ingredients = ["cheese", "eggs", "chicken", "beef", "lamb"]
   random_item = random.choice(random_ingredients)
   print(random_item)
   recipe_results()
elif thinking_of_ingredient == "y":
    print('Okay, we will pick an ingredient in the next few steps')
    recipe_results()
else:
    print('Oops- I didn’t understand that')

#Step six, final step, we use a while loop and booleans to determine if you want to search for another recipe or not 
is_continue = True
while (is_continue == True):
        continue_request = input("Do you want to search for some other ingredient? (Y/N) ")
        if (continue_request.lower() == 'y'):
            is_continue = True
            recipe_results()
        elif (continue_request.lower() == 'n'):
            print("Hope you have found what you need. See you next time!")
            is_continue = False
        else:
            is_continue = True
            print("Invalid response, please try again...")
```
